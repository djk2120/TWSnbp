```python
import xarray as xr
import numpy as np
import pandas as pd
```

```python
#create dummy climate data
p=xr.DataArray(np.arange(150).reshape([15,10]),dims=['lat','lon'])
t=xr.DataArray(np.arange(150).reshape([10,15]).T,dims=['lat','lon'])
ds=xr.Dataset({'T':t,'P':p})
ds
```

```python
#create dummy zones
zones=pd.DataFrame({'P':np.random.randint(150,size=10),'T':np.random.randint(150,size=10)})
zones
```

```python
#convert to xarray dataset
zones.index.name='zone'
ds_zones=xr.Dataset(zones)
ds_zones
```

```python
#compute euclidean distance
euc=(ds.P-ds_zones.P)**2+(ds.T-ds_zones.T)**2
euc.argmin(dim='zone')


```
